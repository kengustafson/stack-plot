# Stack plot example using random numbers

```python
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Create some random data to plot

```python
x = range(1,21)
y1 = np.random.rand(20)
y2 = np.random.rand(20)

print(x)
print(y1)
print(y2)
```

```python
# setup plot legend
plt.plot([], [], color='r', label='y1')
plt.plot([], [], color='b', label='y1+y2')

# setup the plot
plt.stackplot(x, y1, y2, colors=['r', 'b'])
plt.title('Stack Plot Title!')
plt.legend()
plt.show()
```
